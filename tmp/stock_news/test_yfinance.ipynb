```python
import pandas as pd
import numpy as np

# Assuming you have a DataFrame `df` with columns 'Date' and 'Close' (daily closing prices)
# Example:
# df = pd.DataFrame({'Date': [...], 'Close': [...]})
df = hist_3mo.copy()

def calculate_sma(df, window=20):
    """Calculate Simple Moving Average."""
    df['SMA'] = df['Close'].rolling(window=window).mean()
    return df

def calculate_rsi(df, window=14):
    """Calculate Relative Strength Index (RSI) with exponential moving averages to avoid NaN issues."""
    delta = df['Close'].diff(1)
    
    # Separate gains and losses
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    # Calculate exponential moving averages for gain and loss
    avg_gain = gain.ewm(com=window - 1, min_periods=1).mean()
    avg_loss = loss.ewm(com=window - 1, min_periods=1).mean()
    
    # Calculate RS (Relative Strength) and RSI
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    
    return df

def generate_buy_signals(df, sma_window=20, rsi_window=14, rsi_buy_threshold=40):
    """Generate buy signals based on SMA crossover and RSI threshold."""
    df = calculate_sma(df, sma_window)
    df = calculate_rsi(df, rsi_window)
    buy_signals = []
    
    for i in range(1, len(df)):
        # Buy condition: Price crosses above SMA and RSI is below buy threshold
        if df['Close'].iloc[i] > df['SMA'].iloc[i] and df['RSI'].iloc[i] < rsi_buy_threshold:
            buy_signals.append((df['Date'].iloc[i], df['Close'].iloc[i]))

    return buy_signals

# Example usage
# Load your data into df
# df = pd.read_csv("your_stock_data.csv")
buy_signals = generate_buy_signals(df)
print("Buy Signals:")
for date, price in buy_signals:
    print(f"Date: {date}, Buy Price: {price}")
```
